Как мы уже выяснили, перед нами стоит задача memory-based рекомендаций, т к в датасете содержатся только данные о взаимодействии пользователей с товарами и время, без каких-либо метаданных. Также помним о том, что фидбек пользователей неявный, потому что нет оценок товаров, есть только факт покупки.  
Проанализируем имеющиеся решения.

Collaborative Filtering  
Подход основывается на сходстве между пользователями или предметами для предсказания взаимодействий. User-based CF использует схожесть предпочтений пользователей, чтобы рекомендовать предметы, которые понравились другим похожим пользователям. Item-based CF, напротив, использует схожесть между самими предметами для предсказания предпочтений пользователей. Для вычисления сходства обычно используются различные метрики, такие как косинусное сходство или корреляция. Преимущества:  простота реализации, интерпретируемость, вариативность. Недостатки: проблема "холодного старта", сложность нахождения соседей в разреженной матрице, сложность масштабирования, проблема "блуждания" (в некоторых случаях пользователи могут предпочитать предметы, которые не совсем соответствуют их истинным интересам или предпочтениям)

Matrix factorization  
Простая модель embedding'ов. Подход основан на разложении матрицы взаимодействий пользователей-товаров на более низкоранговые матрицы embedding'ов пользователей и товаров с целью заполнить пропущенные значения в исходной матрице. Каждый пользователь и каждый предмет описывается некоторым набором латентных факторов. Преимущества: позволяет работать с разреженными данными, учитывает скрытые взаимодействия. Недостатки: не может работать с новыми товарами и пользователями (холодный старт), не учитывает временную динамику, может быть вычислительно сложным при работе с большими матрицами взаимодействий.  
Есть разные методы выбора objective function для разложения матрицы - SVD, Weighted Matrix Factorization, Matrix Factorization Neural Networks. Последний - комбинация классической матричной факторизации с нейронными сетями. Модель MFNN использует нейронную сеть для предсказания взаимодействий между пользователями и предметами на основе скрытых факторов. Она позволяет учитывать нелинейные зависимости между факторами и повышает точность предсказаний. Преимущества: устойчивость к разреженности данных, обобщение на новых пользователей и предметы, учет линейных и нелинейных аспектов. Недостатки: вычислительная сложность, проблемы с масштабируемостью.  
Factorization Machines with Temporal Dynamics  - метод рекомендательных систем, который комбинирует факторные машины (Factorization Machines) с учетом временной динамики данных. Он расширяет классическую модель факторных машин, предназначенную для работы со статическими данными, и учитывает изменения в предпочтениях пользователей и свойствах предметов со временем. Это достигается путем добавления временной компоненты к модели, которая позволяет моделировать предпочтения пользователей и свойства предметов со временем. Преимущества: учет временной динамики, моделирование сложных взаимодействий. Недостатки: вычислительная сложность, ограниченность объема данных

Sequental Recommendation  
Подход относится к задаче рекомендации, где требуется предлагать элементы пользователю в определенном порядке или последовательности. Принцип работы последовательных рекомендаций часто основан на анализе пользовательской истории, где представлены предыдущие взаимодействия с элементами. На основе этой информации и текущего контекста, такого как время, местоположение или сценарий использования, система рекомендаций старается предсказать, какие элементы могут быть наиболее интересными или релевантными в данный момент или в будущем. Преимущества: учет зависимостей, улучшение долгосрочных интересов. Недостатки: холодный старт, сложность вычислений.  
Традиционные методы - это Марковские модели (Markov Models) и Рекуррентные нейронные сети (Recurrent Neural Networks). Цепи Маркова предполагают, что следующее действие пользователя можно предсказать на основе только его последних (или нескольких последних) действий, в то время как RNN в принципе позволяют раскрыть долгосрочную семантику. SASRec балансирует эти две цели, позволяя фиксировать долгосрочную семантику (как RNN), но, используя механизм внимания, делает свои прогнозы на основе относительно небольшого количества действий (как MC).

Beyond Self-Attention for Sequential Recommendation (BSARec).  
Модели последовательных рекомендаций (SR), основанные на трансформерах, добились блестящих успехов. BSARec - один из более новых и сложных подходов, использующих механизм внимания, выявляет зависимости по последовательности взаимодействий. Этот подход позволяет выделять сложные закономерности подобно RNN-моделям, но снизить требования к объемам обучающей выборки. Решает 2 проблемы предыдущих Sequential Recommendation моделей - недостаточное индуктивное смещение и чрезмерное сглаживание частот. С помощью Фурье преобразования, BSARec получает доступ к индуктивному смещению частот, позволяя улавливать важные закономерности и периодичность, которые можно не заметить, используя только self-attention механизм трансформеров. В архитектуру введён преобразователь частоты, который может захватывать как высокочастотные модели поведения, обусловленные краткосрочными тенденциями, так и низкочастотные, обусловленные долгосрочными интересами, в поведенческих моделях пользователя. Недостатки: вычислительная сложность, ограниченность объема данных, недостаток интерпретируемости

Для данной задачи стоит использовать методы, учитывающие сложные зависимости, временную динамику и устойчивые к разреженным данным. Думаю, больше всего подойдёт BSARec, потому что эта модель заточена именно под последовательность взаимодействий, когда FMTD учитывает именно диманику признаков товаров, которых в нашем датасете нет. Можно также использовать BSARec в сочетании с подходом MFNN, чтобы первый выявлял зависимости по последовательности взаимодействий, а второй выявлял зависимости между скрытыми признаками товаров и покупателей.
